```python
from pathlib import Path
import numpy as np

import matplotlib.pyplot as plt
```

```python
model_identifier = 'bi_lambda-30_lb-10'
compression_folder = Path(f'compression/{model_identifier}')
fnames = list(compression_folder.glob('wedge*npz'))
```

```python
fname = fnames[2]
with np.load(fname) as handle:
    data = handle['data'].squeeze()
    prob = handle['prob'].squeeze()
    regr = handle['regr'].squeeze()
```

```python
occupancy = (data > 0).sum() / np.prod(data.shape)
print(occupancy)
prob_signal = prob[data > 0]

fig, ax = plt.subplots(1, 1)
ax.set_yscale('log')
hist = ax.hist(prob_signal, bins=100)
```

```python
y, x = hist[:2]
x = (x[1:] + x[:-1]) / 2
y_cum = np.cumsum(y)
plt.plot(x, y_cum)
```
